# 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)
and make data frame.

```python
page1=requests.get('https://www.imdb.com/chart/top/?ref_=nv_mv_250')
page1
```

```python
soup1=BeautifulSoup(page1.content)
#soup1
```

```python
#scrapping all movie names
names=[]
movies = soup1.select('td.titleColumn')
for i in range(0,100):
    movie_string=movies[i].get_text()
    movie = (' '.join(movie_string.split()).replace('.', ''))
    movie_title = movie[len(str(i))+1:-7]
    names.append(movie_title)
#names
```

```python
#scrapping all ratings
rating=[]
for i in soup1.find_all('td',class_="ratingColumn imdbRating"):
    rating.append(i.text.replace('\n',''))
rating=rating[0:100]
#ratings
```

```python
soup2=BeautifulSoup(page1.content)
#soup2
```

```python
#scrapping all release years
year=[]
for i in soup2.find_all('span',class_="secondaryInfo"):
    year.append(i.text.replace('(','').replace(')',''))
year=year[0:100]
#year
```

```python
#making DataFrame
import pandas as pd
df=pd.DataFrame({'movie_name':names,'rating':rating,'year':year})
df.index=df.index+1
df
```

# 9) Write a python program to scrape mentioned details from dineout.co.in :
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
page13=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page13
```

```python
soup13=BeautifulSoup(page13.content)
#soup13
```

```python
#scraping restaurant names
name=[]
for i in soup13.find_all('div',class_="restnt-info cursor"):
    name.append(i.text)
name 

#scraping locations
locations=[]
for i in soup13.find_all('div',class_="restnt-loc ellipsis"):
    locations.append(i.text)
locations

#scraping cuisine
cuisine=[]
for i in soup13.find_all('span',class_="double-line-ellipsis"):
    cuisine.append((i.text.replace('₹','')[21:-1]).replace('|','').replace(' ',''))
cuisine

#scraping rating
rating=[]
for i in soup13.find_all('div',class_="restnt-rating rating-4"):
     rating.append(i.text)
rating
#scraping image url
images=[]
for i in soup13.find_all('img',class_="no-img"):
    images.append(i['data-src'])
images

import pandas as pd
df=pd.DataFrame({'Restaurant Name':name,'Cuisine':cuisine,'Location':locations,'Rating':rating,'Image URL':images})
df
```
