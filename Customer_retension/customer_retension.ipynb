```python
from sklearn.preprocessing import OrdinalEncoder
enc=OrdinalEncoder()
```

```python
for i in df.columns:
    if df[i].dtypes=='object':
        df[i]=enc.fit_transform(df[i].values.reshape(-1,1))
```
